# Introducion

### Problem Statement:
Cancer is a leading cause of death worldwide, and early detection is essential for improving survival rates. Cancer is a disease in which cells grow uncontrollably and spread to other parts of the body. It can be classified into two broad categories: malignant and benign.


*   **Malignant** cancer is cancerous, meaning that it can invade and spread to other parts of the body. It is the most common type of cancer and is responsible for most cancer deaths.
*   **Benign** cancer is non-cancerous, meaning that it cannot invade or spread to other parts of the body. It is less common than malignant cancer and is usually less serious.

### Breakdown of the Problem Statement:
* Supervised machine learning problem.
* The target value will be `Diagnosis`.

### Aim of the NoteBook:
This machine learning model aims to predict malignant and benign cancer in early stages. This could improve cancer detection rates and survival outcomes.

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
newdf = data.select_dtypes(include=numerics)
```

```python
for col in newdf.columns:

  #BoxPlot before removing outliers
  fig, ax = plt.subplots(nrows= 1, ncols=2)
  fig.set_size_inches(12, 7)
  ax[0].set_title("[{}] Before Removing Outliers".format(str(col)))
  sns.boxplot(data=data[col],orient="v", color = 'c', ax=ax[0])

  #Removing Outliers
  Q1= data[col].quantile(0.25)
  Q3= data[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-(1.5*IQR)
  upper_bound=Q3+(1.5*IQR)
  data[col]=np.where(data[col]>upper_bound,upper_bound,data[col])
  data[col]=np.where(data[col]<lower_bound,lower_bound,data[col])

  #BoxPlot After Removing Outliers
  ax[1].set_title("[{}] After Removing Outliers".format(str(col)))
  sns.boxplot(data=data[col],orient="v", color = 'c', ax=ax[1])
  plt.show()
```

Data Transformation

```python
diagnosis_cat = {
    'M': 0,
    'B': 1
}
```

```python
data['diagnosis'] = data['diagnosis'].map(diagnosis_cat)
y = data['diagnosis']
x = data.drop(['diagnosis'], axis=1)
```

Standardize the data so that the mean of each variable is 0 and the standard deviation is 1.
Standardization assigns equal weights/importance to each feature, ensuring that no one factor biases model performance in one way just because it is more valuable.

```python
X = StandardScaler().fit_transform(x)
```

## **Feature Selection**

Using Recursive Forest Elimination for feature selection.

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Create an RFE object
rfe = RFE(estimator=RandomForestClassifier(), n_features_to_select=10)
X=X=pd.DataFrame(X)
# Fit the RFE object to the data
rfe.fit(X, y)

# Select the most important features
selected_features = X.columns[rfe.get_support()]

# Create a new DataFrame with the selected features
X = X[selected_features]

# Save the transformed data
X.to_csv('transformed_data.csv')

# Split the data into training and test sets
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```
